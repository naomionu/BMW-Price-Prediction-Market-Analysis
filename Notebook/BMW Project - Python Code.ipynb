```python
# BMW used car market analysis
#-----------------------------------------------
# By: Naomi Onu
# Tools: Python and Power BI
# Dataset: Used BMW listings sourced from Kaggle
     # indicated in README 
#-----------------------------------------------
# Objective: identify the key factors driving used BMW resale prices and evaluate whether a machine learning model can
#            accurately predict a vehicle's price from its characteristics.
```

```python
# Imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
# Load data
df = pd.read_csv("bmw_dataset.csv")
print("Dataset info: ")
df.head()
```

```python
# Initial data exploration
print("Data information: ")
print(df.info())
print()
print("Data description: ")
print(df.describe())
print()
print("Data shape: ")
print(df.shape)
print()
print("Data columns: ")
print(df.columns)
print()
```

```python
# Data cleaning and feature engineering

# Convert units (1 GDP =~ 1.85 CAD, 1 mile =~ 1.61 km, and MPG to L/100km)
df["price_cad"] = df["price"] * 1.85
df["mileage_km"] = df["mileage"] * 1.61
df['fuel_consumption'] = 235.215 / df['mpg']

# New variables
df["car_age"] = 2026 - df["year"]
df["price_per_km"] = np.where(df["mileage_km"] > 0, df["price_cad"] / df["mileage_km"], np.nan)

# Check category values
print()
print("Categories")
print("Transmissions: ", df["transmission"].unique())
print("Fuel types: ", df["fuelType"].unique())
print("Models: ", df["model"].unique())

# Flag unrealistic fuel consumption values
fuel_consumption_outliers = df[(df["fuel_consumption"] <= 2) | (df["fuel_consumption"] >= 25)]
print()
print("Fuel consumption outliers: ")
print(fuel_consumption_outliers)
          
# Remove unrealistic fuel consumption values as these are electric/hybrid vehicles where MPG figures are not comparable to combustion engines.
df = df[(df["fuel_consumption"] > 2) & (df["fuel_consumption"] < 25)]

# Remove impossible values
df = df[df["price_cad"] > 0]
df = df[df["mileage_km"] >= 0]
df = df[df["car_age"] >= 0]
```

```python
# Data exploration

print("Number of listings by BMW model: ")
print(df["model"].value_counts())
print()

print("Average vehicle price: $", df["price_cad"].mean())
print()

print("Average price per model (CAD): ")
print(df.groupby("model")["price_cad"].mean().sort_values(ascending = False))
print()

print("Average mileage by model (km): ")
print(df.groupby("model")["mileage_km"].mean().sort_values(ascending = False))
print()

print("Average price by transmission (CAD): ")
print(df.groupby("transmission")["price_cad"].mean().sort_values(ascending = False))
print()

print("Average price by fuel type (CAD): ")
print(df.groupby("fuelType")["price_cad"].mean().sort_values(ascending = False))
```

```python
# Graph 1
plt.hist(df["price_cad"], bins = 30)
plt.xlabel("Price (CAD)")
plt.ylabel("Count")
plt.title("Distribution of BMW prices")
plt.show()
```

```python
# Graph 2
sns.scatterplot(x = "mileage_km",
                y = "price_cad",
                data = df,
                hue = "fuelType",
                s = 10)
plt.xlabel("Mileage (km)")
plt.ylabel("Price (CAD)")
plt.title("Mileage vs price by fuel type")
plt.show()
```

```python
# Graph 3
plt.figure(figsize = (8, 4))
sns.regplot(x = "mileage_km",
            y = "price_cad",
            data = df,
            scatter_kws = {"s": 5},
            line_kws = {"color": "red"})
plt.xlabel("Mileage (km)")
plt.ylabel("Price (CAD)")
plt.title("Mileage vs. price with trend line")
plt.show()
```

```python
# Graph 4
plt.figure(figsize = (10, 4))
sns.boxplot(x = "model",
            y = "price_cad",
            data = df)
plt.xticks(rotation = 45)
plt.xlabel("BMW model")
plt.ylabel("Price (CAD)")
plt.title("Price distribution by BMW model")
plt.show()
```

```python
# Graph 5
avg_price_model = (df.groupby("model")["price_cad"].mean().sort_values(ascending=False))
plt.figure(figsize = (10, 4))
avg_price_model.plot(kind = "bar")
plt.ylabel("Price (CAD)")
plt.xlabel("Model")
plt.title("Average price by BMW model")
plt.xticks(rotation = 45)
plt.show()
```

```python
# Graph 6
avg_price_age = (df.groupby("car_age")["price_cad"].mean().reset_index())
plt.figure(figsize = (8, 4))
sns.lineplot(x = "car_age",
             y = "price_cad",
             data = avg_price_age,
             marker = "o")
plt.xlabel("Car age")
plt.ylabel("Price (CAD)")
plt.title("Depreciation curve: average price by car age")
plt.show()
```

```python
# Graph 7
plt.figure(figsize = (4, 4))
sns.barplot(x = "transmission",
            y = "price_cad",
            data = df)
plt.xlabel("Transmission")
plt.ylabel("Price (CAD)")
plt.title("Average price by transmission type")
plt.show()
```

```python
# Graph 8
numeric_cols = ["year", "price_cad", "mileage_km", "tax", "fuel_consumption", "engineSize", "car_age", "price_per_km"]
corr = df[numeric_cols].corr()
plt.figure(figsize = (10, 4))
sns.heatmap(corr, 
            annot = True,
            cmap = "coolwarm",
            fmt = ".2f")
plt.xticks(rotation = 45)
plt.title("Correlation heatmap")
plt.show()
```

```python
# Linear regression modeling
features = ["mileage_km", "car_age", "engineSize", "fuel_consumption", "tax"]
target = "price_cad"

x = df[features]
y = df[target]

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)
linear_model = LinearRegression()
linear_model.fit(x_train, y_train)
linear_predictions = linear_model.predict(x_test)
linear_rmse = round(np.sqrt(mean_squared_error(y_test, linear_predictions)), 2)
linear_mae = round(mean_absolute_error(y_test, linear_predictions), 2)
linear_r2 = round(r2_score(y_test, linear_predictions), 4)

print("Linear regression results: ")
print("Root mean square error (RMSE): ", linear_rmse)
print("Mean absolute error (MAE): ", linear_mae)
print("Coefficient of determination (R^2): ", linear_r2)
print()

linear_coefficients = pd.DataFrame({"Feature": features, "Coefficient": linear_model.coef_})
print(round(linear_coefficients, 2))
print()

plt.figure(figsize = (8, 4))
plt.scatter(y_test, linear_predictions, s = 10)
plt.xlabel("Actual price (CAD)")
plt.ylabel("Predicted price (CAD)")
plt.title("Linear regression: actual vs. predicted prices")
plt.show()
```

```python
# Random forest modeling
rf_model = RandomForestRegressor(n_estimators = 100, random_state = 42)
rf_model.fit(x_train, y_train)
rf_predictions = rf_model.predict(x_test)
rf_rmse = round(np.sqrt(mean_squared_error(y_test, rf_predictions)), 2)
rf_mae = round(mean_absolute_error(y_test, rf_predictions), 2)
rf_r2 = round(r2_score(y_test, rf_predictions), 4)

print("Random forest results: ")
print("Root mean square error (RMSE):", rf_rmse)
print("Mean absolute error (MAE):", rf_mae)
print("Coefficient of determination (R^2):", rf_r2)
print()

feature_importance = pd.DataFrame({"Feature": features, "Importance": rf_model.feature_importances_}).sort_values(by = "Importance", ascending = False)
print(round(feature_importance, 4))
print()

plt.figure(figsize = (6, 2))
sns.barplot(x = "Importance", y = "Feature", data = feature_importance)
plt.title("Random forest feature importance")
plt.xlabel("Importance")
plt.ylabel("Feature")
plt.show()
print()

plt.figure(figsize = (8, 4))
plt.scatter(y_test, rf_predictions, s = 10)
plt.xlabel("Actual price (CAD)")
plt.ylabel("Predicted price (CAD)")
plt.title("Random forest: actual vs predicted prices")
plt.show()
```

```python
# Linear regression vs. random forest comparison
model_results = pd.DataFrame({"Model": ["Linear regression", "Random forest"], "RMSE": [linear_rmse, rf_rmse], "MAE": [linear_mae, rf_mae], "R^2": [linear_r2, rf_r2]})
print(model_results)
print()
print("R^2 improvement: ", round(rf_r2 - linear_r2, 4))

plt.figure(figsize = (4, 4))
sns.barplot(x = "Model", y = "R^2", data = model_results)
plt.title("Model comparison by R^2")
plt.ylabel("R^2 Score")
plt.show()
```

```python
# Actual vs. predicted scatterplot for top 5 models
predictions_df = pd.DataFrame({"Actual": y_test, "Predicted": rf_predictions})
predictions_df["model"] = df.loc[y_test.index, "model"]

top_models = (df["model"].value_counts().head(5).index)
filtered_df = predictions_df[predictions_df["model"].isin(top_models)]

plt.figure(figsize = (8,4))
sns.scatterplot(data = filtered_df, x = "Actual", y = "Predicted", hue = "model", s = 40)

max_price = max(filtered_df["Actual"].max(), filtered_df["Predicted"].max())
min_price = min(filtered_df["Actual"].min(), filtered_df["Predicted"].min())

plt.plot([min_price, max_price], [min_price, max_price], color = "red", linestyle = "--", label = "Perfect Prediction")
plt.xlabel("Actual Price (CAD)")
plt.ylabel("Predicted Price (CAD)")
plt.title("Actual vs. Predicted BMW Prices")
plt.legend()
plt.show()
```

```python
# Cleaned datasets exported as a csv for use in the Power BI dashboard
# Cleaned original dataset
df.to_csv("cleaned_bmw_data.csv", index = False)
print("Cleaned dataset exported as cleaned_bmw_data.csv")

# Random forest feature importance
feature_importance.to_csv(feature_importance.csv", index=False)
print("Feature importance dataset exported as feature_importance.csv")
                          
# Actual vs. predicted data for top 5 models
predictions_df = pd.DataFrame({"actual_price_cad": y_test.values, "predicted_price_cad": rf_predictions}, index=y_test.index)
predictions_df["model"] = df.loc[y_test.index, "model"].values
top_models = (df["model"].value_counts().head(5).index)
predictions_powerbi = predictions_df[predictions_df["model"].isin(top_models)].copy()
predictions_powerbi["listing_id"] = range(1, len(predictions_powerbi) + 1)
predictions_powerbi.to_csv("actual_vs_predicted_top5.csv", index=False)
print("Actual vs. predicted dataset exported as actual_vs_predicted_top5.csv")
```
